__1.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### $$\mbox{arctg}(\frac{y}{x})=\ln \sqrt{x^2+y^2}$$
### $$\frac{(\frac{y}{x})'}{1+(\frac{y}{x})^2}=\frac{(\sqrt{x^2+y^2})'}{\sqrt{x^2+y^2}}$$
### $$\frac{\frac{y'x-y}{x^2}}{1+(\frac{y}{x})^2}=\frac{\frac{(x^2+y^2)'}{2\sqrt{x^2+y^2}}}{\sqrt{x^2+y^2}}$$
### $$\frac{y'x-y}{x^2+y^2}=\frac{(x^2+y^2)'}{2\cdot(x^2+y^2)}$$
### $$\frac{y'x-y}{x^2+y^2}=\frac{2x+2yy'}{2\cdot(x^2+y^2)}$$
### $$y'x-y=x+yy'$$
### $$y'x-yy'=x+y$$
### $$y'(x-y)=y+x$$
### $$y'=\frac{x+y}{x-y}$$

__2.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### \begin{cases}
   y=\frac{t^2}{t-1}, \\
   x=\frac{t}{t^2-1} 
 \end{cases}

###  $$y'_{x}=\frac{y'_{t}}{x'_{t}}=\frac{(\frac{t^2}{t-1})'}{(\frac{t}{t^2-1} )'}=
\frac{\frac{2t(t-1)-t^2}{(t-1)^2}}{\frac{t^2-1-2t^2}{(t^2-1)^2}}=
-\frac{(t^2-2t)(t^2-1)^2}{(t^2+1)(t-1)^2}
 $$

__3.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è:

### $$y=(x^2+2)^5\cdot(3x-x^3)^3$$

### $$y'=(x^2+2)^5\cdot(3x-x^3)^3\cdot(\ln((x^2+2)^5\cdot(3x-x^3)^3))'=
(x^2+2)^5\cdot(3x-x^3)^3\cdot
(\frac{5(x^2+2)'}{(x^2+2)}+\frac{3(3x-x^3)'}{(3x-x^3)})=
(x^2+2)^5\cdot(3x-x^3)^3\cdot
(\frac{10x}{(x^2+2)}+\frac{9-9x^2}{(3x-x^3)})=

(x^2+2)^4\cdot(3x-x^3)^3\cdot10x+(x^2+2)^5\cdot(3x-x^3)^2\cdot(9-9x^2)
$$

__4.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Ñ—É–Ω–∫—Ü–∏–∏ —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è:

### $$y=x^x$$

### $$y'=x^x\cdot(\ln(x^x))'
=x^x\cdot(x\cdot\ln(x))'=
x^x\cdot(1\cdot\ln(x)+x\cdot\frac{1}{x})=
x^x\cdot(\ln(x)+1)$$

__5.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Ñ—É–Ω–∫—Ü–∏–∏ —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è:

### $$y=\frac{(2-x^2)^3\cdot(x-1)^2}{(2x^3-3x)\cdot e^x}$$
